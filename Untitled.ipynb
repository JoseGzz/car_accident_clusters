```python
import pandas as pd
import numpy as np
```

```python
data_raw = pd.read_csv('rativ_abierto_2025.csv')
data_raw.head()
```

```python
data = data_raw[['Día', 'Mes', 'Año', 'Referencia']]

data["date"] = pd.to_datetime(
    data["Mes"].astype(str) + "-" +
    data["Día"].astype(str) + "-" +
    data["Año"].astype(str)
)
```

```python
data_final = data[['date', 'Referencia']]
data_final.columns = ['date', 'coords']

# Replace 99 with NaN first
data_final["coords"] = data_final["coords"].replace("99", np.nan)

# Extract latitude (first number)
data_final["latitude"] = data_final["coords"].str.extract(r'^\s*([+-]?\d+\.\d+)\s*,')
# Extract longitude (second number)
data_final["longitude"] = data_final["coords"].str.extract(r',\s*([+-]?\d+\.\d+)\s*$')

# Convert to float
data_final["latitude"] = data_final["latitude"].astype(float)
data_final["longitude"] = data_final["longitude"].astype(float)
data_final = data_final.dropna()

data_final = data_final.drop('coords', axis=1)[['latitude','longitude','date']]
data_final.head()
```

```python
data_final.to_csv('accidents.csv', index=False)
```
